````python
from prettytable import PrettyTable

# Lista para almacenar las semirreacciones ingresadas
datos_redox = {}

# Potenciales de reducción de los electrodos de sacrificio
potencial_mg = -2.38
potencial_zn = -0.76
potencial_al = -1.66

# Solicitar semirreacciones al usuario hasta que se ingrese el número 0
while True:
    semirreaccion = input("Ingrese una semirreacción (o 0 para salir): ")

    if semirreaccion == '0':
        break

    # Obtener el potencial de reducción de la semirreacción
    potencial = datos_redox.get(semirreaccion)

    if potencial is None:
        # Si la semirreacción no está en la lista, solicitar el potencial
        potencial = float(input("Ingrese el Potencial de Reducción (E°): "))

    # Almacenar la semirreacción y su potencial en el diccionario
    datos_redox[semirreaccion] = potencial

# Crear la tabla con los datos redox y la nueva columna
tabla_datos = PrettyTable()
tabla_datos.field_names = ["Semi - Reacciones", "Potencial de Reducción (E°)", "Posibles electrodos de Sacrificio"]

# Determinar los electrodos de sacrificio aplicables a todas las semirreacciones
electrodos_comunes = set(["Mg", "Zn", "Al"])

for reaccion, potencial in datos_redox.items():
    # Determinar los electrodos de sacrificio aplicables a la semirreacción actual
    electrodos_sacrificio = [elec for elec, pot in [("Mg", potencial_mg), ("Zn", potencial_zn), ("Al", potencial_al)] if pot < potencial]

    # Agregar la fila a la tabla
    tabla_datos.add_row([reaccion, potencial, ", ".join(electrodos_sacrificio)])

    # Actualizar los electrodos comunes
    electrodos_comunes.intersection_update(electrodos_sacrificio)

# Imprimir la tabla
print(tabla_datos)

# Imprimir los electrodos de sacrificio comunes
if electrodos_comunes:
    print("Electrodos de Sacrificio de la Reaccion:", ", ".join(electrodos_comunes))
else:
    print("No hay electrodos de sacrificio comunes.")
````

````python
import json
from prettytable import PrettyTable

# Cargar el diccionario desde el archivo
ruta_archivo = 'datos_redox.json'

with open(ruta_archivo, 'r') as archivo:
    datos_redox = json.load(archivo)

# Potenciales de reducción de los electrodos de sacrificio
potencial_mg = -2.38
potencial_zn = -0.76
potencial_al = -1.66

# Crear la tabla con los datos redox
tabla_datos = PrettyTable()
tabla_datos.field_names = ["Semi - Reacciones", "Potencial de Reducción (E°)"]

# Agregar filas a la tabla
for reaccion, potencial in datos_redox.items():
    tabla_datos.add_row([reaccion, potencial])

# Imprimir la tabla
print(tabla_datos)

# Pedir al usuario el número de semirreacciones a tratar
mensaje = "Hola, no deje espacios despues de copiar la semireacción."
print(mensaje
      )
num_semirreacciones = int(input("Ingrese el número de semirreacciones a tratar: "))

# Lista para almacenar las semirreacciones ingresadas
semirreacciones_ingresadas = []

# Pedir al usuario cada semirreacción
for _ in range(num_semirreacciones):
    semirreaccion = input("Ingrese una semirreacción: ")
    semirreacciones_ingresadas.append(semirreaccion)

# Encontrar el menor Potencial de Reducción (Eo) entre las semirreacciones ingresadas
menor_potencial_ingresado = min(datos_redox[semirreaccion] for semirreaccion in semirreacciones_ingresadas)

# Determinar los electrodos de sacrificio aplicables
electrodos_sacrificio = []

if potencial_mg < menor_potencial_ingresado:
    electrodos_sacrificio.append("Magnesio")
if potencial_zn < menor_potencial_ingresado:
    electrodos_sacrificio.append("Zinc")
if potencial_al < menor_potencial_ingresado:
    electrodos_sacrificio.append("Aluminio")

# Imprimir los electrodos de sacrificio posibles
if electrodos_sacrificio:
    print("Posibles electrodos de sacrificio:", ", ".join(electrodos_sacrificio))
else:
    print("No hay electrodos de sacrificio posibles.")
````
